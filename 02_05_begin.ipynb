### 3.10 Initial Check for Outliers

Identifying outliers should be done during both the initial data check and the data preprocessing stages, but with different purposes and depth of analysis.

Why? We need to understand Data Distribution, so we get a preliminary understanding of how the data is distributed. Also, this initial check allows us to quickly identify any glaring outliers or data entry errors.

We'll use a visualization to check for outliers now. Later, we'll use more code.



```python
# This is where the matplot library is used.
# Run this cell to create a box plot
plt.boxplot(df['Age'])
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

Let's Check all numeric features for outlers - instead of one at a time. To do this, we just need to create a new smaller DataFrame that gets all of the numeric features from our larger DataFrame.


```python
# Create box plots for all numeric features

#First, let's get all of the numeric features from the DataFrame
numeric_df = df.select_dtypes(include=[np.number])

# Create box plots for all numeric features
numeric_df.plot(kind='box', subplots=True, layout=(int(len(numeric_df.columns)/3), 3), figsize=(15, 15), sharex=False, sharey=False)
plt.tight_layout()
plt.show()

```
